# Escalado
Conjunto de datos: iris (incluido en sklearn)
Este conjunto de datos es compacto y perfecto para demostrar técnicas de escalado.

```python
# Importar librerías necesarias
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el conjunto de datos
iris = load_iris(as_frame=True)
data = iris.data
```

```python
# Visualizar las estadísticas iniciales
print("Estadísticas iniciales:")
data.describe()
```

```python
# Aplicar MinMaxScaler
minmax_scaler = MinMaxScaler()
data_minmax = pd.DataFrame(minmax_scaler.fit_transform(data), columns=data.columns)
```

```python
# Aplicar StandardScaler
standard_scaler = StandardScaler()
data_standard = pd.DataFrame(standard_scaler.fit_transform(data), columns=data.columns)
```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
sns.set_theme(style="darkgrid")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Gráfica Antes del Escalado 
sns.histplot(data['sepal length (cm)'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('1. Distribución Original (cm)')
axes[0].set_xlabel('Largo del Sépalo')
axes[0].set_ylabel('Frecuencia')

# 2. Gráfica Después del Escalado
sns.histplot(data_minmax['sepal length (cm)'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('2. Distribución con MinMaxScaler [0, 1]')
axes[1].set_xlabel('Largo del Sépalo (Escalado)')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='sepal length (cm)', 
    y='sepal width (cm)', 
    data=data_standard, 
    hue=iris.target, 
    palette='viridis',
    alpha=0.8
)

plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)

plt.title('Dispersión: Sepal Length vs Sepal Width (Datos Estandarizados)', fontsize=14)
plt.xlabel('Sepal Length (Z-score)')
plt.ylabel('Sepal Width (Z-score)')
plt.grid(True, linestyle=':', alpha=0.6)

plt.show()
```
